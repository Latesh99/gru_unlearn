```python
import numpy as np
import matplotlib.pyplot as plt

# Simulated Data (Replace with actual values if available)
epochs = np.arange(1, 11)  # Assuming 10 epochs for each phase
train_loss = np.random.uniform(0.2, 0.4, 10)  # Random training loss
unlearn_loss = np.random.uniform(0.3, 0.5, 10)  # Higher loss due to unlearning
relearn_loss = np.random.uniform(0.2, 0.35, 10)  # Lower loss after relearning

# Accuracy values from previous output
initial_accuracy = 0.9889
unlearned_accuracy = 0.8869
relearned_accuracy = 0.9898

# Layer-wise distance and activation drift values
layerwise_distance = 8.7547
activation_drift = 61.7683

# Relearning Efficiency
relearning_efficiency = 1.0088

# Plot Training, Unlearning, and Relearning Loss
plt.figure(figsize=(12, 5))
plt.plot(epochs, train_loss, label="Training Loss", marker="o")
plt.plot(epochs, unlearn_loss, label="Unlearning Loss", marker="s")
plt.plot(epochs, relearn_loss, label="Relearning Loss", marker="d")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss Curve During Training, Unlearning, and Relearning")
plt.legend()
plt.grid(True)
plt.show()

# Accuracy Bar Chart
plt.figure(figsize=(8, 5))
plt.bar(["Trained", "After Unlearning", "After Relearning"],
        [initial_accuracy, unlearned_accuracy, relearned_accuracy],
        color=["blue", "red", "green"])
plt.ylabel("Accuracy")
plt.title("Accuracy Before and After Unlearning & Relearning")
plt.ylim(0.85, 1.0)
plt.grid(axis="y")
plt.show()

# Layer-wise Distance and Activation Drift
plt.figure(figsize=(8, 5))
plt.bar(["Layer-wise Distance", "Activation Drift"],
        [layerwise_distance, activation_drift],
        color=["purple", "orange"])
plt.ylabel("Distance")
plt.title("Model Changes Due to Unlearning")
plt.grid(axis="y")
plt.show()

# Relearning Efficiency
plt.figure(figsize=(6, 4))
plt.bar(["Relearning Efficiency"], [relearning_efficiency], color="green")
plt.ylabel("Efficiency")
plt.title("Relearning Efficiency Score")
plt.ylim(0.9, 1.1)
plt.grid(axis="y")
plt.show()
```
